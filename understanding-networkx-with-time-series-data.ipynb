```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')[1]
tickers = tickers.Symbol.tolist()
tickers
```

```python
frames = []

for ticker in tickers:
    var = yf.Ticker(ticker)
    frame = var.institutional_holders
    frame['Comp'] = var.ticker
    frames.append(frame)
```

```python
all_tog = pd.concat(frames)
all_tog
```

```python
G = nx.from_pandas_edgelist(all_tog, 'Holder', 'Comp', edge_attr=True)
```

```python
edgelist = nx.to_edgelist(G)
edgelist
```

```python
colors = []

for node in G:
    if node in all_tog['Comp'].values:
        colors.append('red')
    else:
        colors.append('yellow')
```

```python
plt.figure(figsize=(50, 40))
nx.draw(G, with_labels=True, 
        node_color=colors, 
        node_size=[v * 100 for v in dict(G.degree()).values()], # Resizing the nodes based on how many it is connected to.
        width=[v[2]['Value']/50_000_000_000 for v in edgelist]) # Spreading the edge out based on its value
ax = plt.gca()
ax.collections[0].set_edgecolor('#696969') # Gray color
```
